## Keyword analysis

Based on Web of Science searches make keyword analysis

```python
import seaborn as sns
import textwrap
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfKeyword = pd.read_csv('data/resilience_metrics.csv')
dfKeyword.groupby('type').sum()
```

```python
plt.figure(figsize=(12, 4))
dfKeywordCategory = dfKeyword[dfKeyword['type'] == 'category']

ax = sns.barplot(data=dfKeywordCategory, x="keyword", y="number_results")
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height + 20, round(height), ha="center")
    
ax.set_ylim([0,690])
ax.set_xticklabels([textwrap.fill(e, 15) for e in dfKeywordCategory["keyword"]])
ax.set_xlabel('')
ax.set_ylabel('Number of results')
ax.set_title('Abstract Search in WoS - metric category keywords')

plt.show()
```

```python
plt.figure(figsize=(17, 4))
dfKeywordMetric = dfKeyword[dfKeyword['type'] == 'metric']

ax = sns.barplot(data=dfKeywordMetric, x="keyword", y="number_results")
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height + 10, round(height), ha="center")

ax.set_ylim([0,390])
ax.set_xticklabels([textwrap.fill(e, 10) for e in dfKeywordMetric["keyword"]])
ax.set_xlabel('')
ax.set_ylabel('Number of results')
ax.set_title('Abstract Search in WoS - metric keywords')

plt.show()
```

```python
plt.figure(figsize=(17, 4))
dfKeywordOther = dfKeyword[dfKeyword['type'] == 'other'].sort_values('number_results', ascending = False)

ax = sns.barplot(data=dfKeywordOther, x="keyword", y="number_results")
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height + 50, round(height), ha="center")


ax.set_ylim([0,1590])
ax.set_xticklabels([textwrap.fill(e, 15) for e in dfKeywordOther["keyword"]])
ax.set_xlabel('')
ax.set_ylabel('Number of results')
ax.set_title('Abstract Search in WoS - other keywords')

plt.show()
```
